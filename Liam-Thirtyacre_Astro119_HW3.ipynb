```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(0, 2*math.pi, .0062831853071796) #create array x = (0-2pi)
                                                #with 1000 values
a = np.cos(2*x)*5.5+5.5
b = 0.02**x
y = 0.25*x**2 + np.sin(10*x)*0.1
plt.plot(x,y,a,b)
plt.ylabel('Measures of Awesomeness')
plt.xlabel('Time in Astro 119')
plt.xlim(0, 2*math.pi)
plt.ylim(-1, 10)
plt.show()
```
